##### PyMongo
* [PyMongo](https://pypi.org/project/pymongo/)

```python
#Importing packages
# import pymongo # Option 1
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

## Declare Mongo Client

```python
# client = pymongo.MongoClient("mongodb://localhost:27017/") Option 1
client = MongoClient(port=27017) # Option 2
```

## List databases

```python
print(client.list_database_names())
```

## Connect to the 'test' database

```python
db=client.test
```

## Check server status

```python
serverStatusResult=db.command("serverStatus")
```

## List collection names (table names)

```python
db.list_collection_names()
```

## Select one element from the collection where language is turkish

```python
result = db.data.find_one({"lang":"tr"})
```

```python
pprint(result)
```

## Nested Query Select all tweets by Galatasaray SK

```python
result = db.data.find({"user.name":"Galatasaray SK"})
```

```python
for i in result:
    pprint(i)
```

## Count of all tweets where language is turkish

```python
db.data.find({"lang":"tr"}).count()
```

## Convert mongo collection into pandas dataframe

```python
df = pd.DataFrame(list(db.data.find()))
```
